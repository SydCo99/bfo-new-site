```python
import pandas as pd
```

```python
df = pd.read_csv("users.csv")
```

```python
df
```

```python
df = df.drop(columns = ["description", "underneath"])
```

```python
df = df.sort_values('name')
```

```python
url_list = df['url'].tolist()
```

```python
url_list
```

```python
url_hyperlink = []

for url in url_list: 
    url = f'<a href="{str(url).strip()}">{url}</a>'
    url_hyperlink.append(url)
```

```python
url_hyperlink
```

```python
df = df.drop(columns = ["url"])
```

```python
url_hyperlink
```

```python
df["url"] = url_hyperlink
```

```python
df
```

```python
df.to_csv("users_cleaned.csv", index = False)
```
